#Statistics Basics

#Q1.What is statistics, and why is it important?

- Statistics is the branch of mathematics that deals with collecting, organizing, analyzing, and interpreting data.
It helps us make decisions based on data.

#Example:

 A company uses statistics to understand customer preferences and improve products.


#Q2.What are the two main types of statistics?

 - Descriptive Statistics: Summarizes data (mean, median, etc.).

 - Inferential Statistics: Makes predictions or generalizations about a population based on a sample.


#Q3.What are descriptive statistics?

 Descriptive statistics describe and summarize data using numbers like mean, median, mode, etc.

 #Example:
     Average score of a class in a test.


#Q4.What is inferential statistics?

 - Inferential statistics use data from a sample to make conclusions about a larger population.

 #Example:
Surveying 100 people to predict election results for an entire country.


#Q5.hat is sampling in statistics?

 - Sampling is selecting a subset of individuals from a population to estimate characteristics of the whole group.

 #Example:
Choosing 50 students from a school to find out the average study time.


#Q5.What are the different types of sampling methods?

- Random Sampling (Equal chance)

- Systematic Sampling (Every k-th element)

- Stratified Sampling (Dividing population into groups and sampling)

- Cluster Sampling (Dividing into clusters and randomly choosing clusters)


#Q7.What is the difference between random and non-random sampling?

- Random Sampling: Every individual has an equal chance.

- Non-random Sampling: Some bias or personal choice is involved.

#Example:
Random: Lottery draw.

Non-random: Interviewing only your friends.


#Q8.Define and give examples of qualitative and quantitative data.
- Qualitative Data: Non-numeric, like colors or names.
#Example: Gender, color of cars.

- Quantitative Data: Numeric data.
#Example: Age, height, income.


#Q9.What are the different types of data in statistics?
- Nominal Data (Names)

- Ordinal Data (Order)

- Interval Data (No true zero)

- Ratio Data (True zero)


#Q10.Explain nominal, ordinal, interval, and ratio levels of measurement.

- Nominal: Categories only (e.g., Gender, City name)

- Ordinal: Ordered categories (e.g., Ranking 1st, 2nd, 3rd)

- Interval: Numeric, but no true zero (e.g., Temperature in Celsius)

- Ratio: Numeric with true zero (e.g., Height, Weight)



#Q11.What is the measure of central tendency?

- It shows the center of a data set: mean, median, and mode.


#Q12.Define mean, median, and mode.

- Mean: Average (Sum of values รท Number of values)

- Median: Middle value when sorted

- Mode: Most frequent value

#Q13.What is the significance of the measure of central tendency?

- It helps to understand the "typical" value in a dataset.

 #Example:
The average salary helps understand income level.

#Q15.What is standard deviation, and why is it important?

- Standard deviation is the square root of variance. It shows the spread of data.

#Importance:
A small standard deviation = data is close to mean,
A large standard deviation = data is spread out.


#Q16.Define and explain the term range in statistics.

- Range = Maximum value - Minimum value

 #Example:
If highest score = 90, lowest = 40, then range = 90 - 40 = 50.


#Q17.What is the difference between variance and standard deviation?

- Variance is in squared units.

- Standard Deviation is in original units (square root of variance).


#Q18.What is skewness in a dataset?

- Skewness measures the asymmetry of a distribution.



#Q19.What does it mean if a dataset is positively or negatively skewed?

- Positively Skewed: Tail on the right (more low values)

- Negatively Skewed: Tail on the left (more high values)


#Q20.Define and explain kurtosis.

- Kurtosis measures the "peakedness" of a distribution.

- High kurtosis: sharp peak

- Low kurtosis: flat peak


#Q21.What is the purpose of covariance?

- Covariance shows how two variables change together.

(A). Positive covariance: Variables move in same direction.

(B). Negative covariance: Variables move in opposite directions.



#Q22.What does correlation measure in statistics?

- Correlation measures both the strength and direction of a relationship between two variables.

Range: -1 to +1.

#Q23.What is the difference between covariance and correlation?

- Covariance: Only direction (+/-)

- Correlation: Direction + strength (standardized between -1 and +1)



#Q24.What are some real-world applications of statistics?

- Medicine (clinical trials)

- Business (market research)

- Sports (player performance analysis)

- Government (census)

```python
#Q4.Implement sampling techniques like random sampling and stratified sampling.
'''
import random
import pandas as pd

# Random Sampling
data = list(range(1, 101))
random_sample = random.sample(data, 10)
print("Random Sample:", random_sample)

# Stratified Sampling
df = pd.DataFrame({
    'Gender': ['Male', 'Female']*50,
    'Score': random.choices(range(0,100), k=100)
})

stratified_sample = df.groupby('Gender', group_keys=False).apply(lambda x: x.sample(5))
print(stratified_sample)
'''
```
